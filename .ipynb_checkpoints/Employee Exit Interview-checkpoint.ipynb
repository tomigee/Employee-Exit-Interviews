# Analyzing Employee Exit Surveys
In this project, we're provided with two datasets containing information surrouding employee exits. The first dataset is from the DETE (Department of Education, Training and Employment), and the second is from the TAFE (Technical and Further Education) Institute in Queensland, Australia. The aim of this analysis is to answer the following questions:
1. Are employees who only worked for the institutes for a short period of time resigning due to some kind of dissatisfaction? What about employees who have been there longer?
2. Are younger employees resigning due to some kind of dissatisfaction? What about older employees?

## 3 - Data Analysis

Recall that the purpose of this project is to answer the following questions:

2. Are younger employees resigning due to some kind of dissatisfaction? What about older employees?

### 3.1 - Question 1

We'll tackle the first question first. To do that, we need to take a deeper look into the `separationType` columns for both datasets. `separtationType` contains information on the reason each employee leaves the company.

```python
dete_survey_updated['separationtype'].value_counts(dropna=False)
```

```python
tafe_survey_updated['separationtype'].value_counts(dropna=False)
```

Looks like there's a couple reasons why someone might leave either of these companies. Let's isolate the "resignation" `separationtype`s since we're trying to investigate only folks who resigned from the companies. But first, let's remove that pesky NaN

```python
dete_resignations = dete_survey_updated.loc[dete_survey_updated['separationtype'].str.contains('Resignation'),:].copy()
tafe_resignations = tafe_survey_updated.dropna().loc[tafe_survey_updated['separationtype'].dropna().str.contains('Resignation'),:].copy()
```

```python
dete_resignations['cease_date'].value_counts(dropna=False)
```

Looks like we have some `NaN`s in the cease_date column. Let's go ahead and remove all rows with that issue.

```python
dete_resignations = dete_resignations.dropna(subset=['cease_date'])
```

```python
dete_resignations['cease_date'].value_counts()
year_pattern = r'(?P<cease_year>20[0-9]{2})'
dete_resignations['cease_date'] = dete_resignations['cease_date'].str.extract(year_pattern).astype('int64')
```

Nothing looks wrong here. All dates look reasonable. Now let's calculate the years of service of each employee from the dete survey, so we can more easily compare with the `institute_service` of the `tafe_resignations` dataset.

```python
dete_resignations['institute_service'] = dete_resignations['cease_date'] - dete_resignations['dete_start_date']
```

Now let's isolate the people who resigned due to dissatisfaction

```python
display(tafe_resignations['Contributing Factors. Dissatisfaction'].value_counts())
display(tafe_resignations['Contributing Factors. Job Dissatisfaction'].value_counts())
```

Convert the value to `True`, `False`, or `NaN` to make it more easily parseable

```python
def update_vals(element):
    if pd.isnull(element):
        return np.nan
    elif element == '-':
        return False
    else:
        return True
        
tafe_resignations['Contributing Factors. Dissatisfaction'] = tafe_resignations['Contributing Factors. Dissatisfaction'].apply(update_vals)
tafe_resignations['Contributing Factors. Job Dissatisfaction'] = tafe_resignations['Contributing Factors. Job Dissatisfaction'].apply(update_vals)
```

```python
dissatisfaction_df = pd.merge(tafe_resignations['Contributing Factors. Dissatisfaction'],tafe_resignations['Contributing Factors. Job Dissatisfaction'],right_index=True,left_index=True)
tafe_resignations['dissatisfied'] = dissatisfaction_df.any(axis=1, skipna=False)
```

The dete survey has already been formatted as boolean for the columns we need to answer whether or not the employee left due to some kind of dissatisfaction. 

```python
dete_dissatisfaction_cols = ['job_dissatisfaction','dissatisfaction_with_the_department','physical_work_environment',
                             'lack_of_recognition','lack_of_job_security','work_location','employment_conditions',
                             'work_life_balance','workload']
dete_resignations['dissatisfied'] = dete_resignations[dete_dissatisfaction_cols].any(axis=1, skipna=False)
```

Now we have successfully created a `dissatisfied` column in both the `tafe_resignation` and `dete_resignation` datasets to indicate whether or not an employee was dissatisfied when they resigned. Now let's merge the dataframes together and then drop columns with a large number of null values (corresponding to columns not common between both datasets)

```python
dete_resignations['institute'] = 'DETE'
tafe_resignations['institute'] = 'TAFE'
combined = pd.concat([dete_resignations,tafe_resignations],axis=0,ignore_index=True)
combined_updated = combined.dropna(axis=1, thresh=500)
# display(dete_resignations.columns)
# display(tafe_resignations.columns)
```

The question we're trying to answer delineates between folks who worked at DETE/TAFE for a short time, vs those who worked for a long time. As such, we need to cluster our population to reflect their years of service/experience. To do so, we'll create three categories of experience: "New", "Experienced", and "Established" to categorize each employee. We'll call this column `service_cat`.

```python
combined_updated['institute_service'] = combined_updated['institute_service'].astype('str').copy()
pattern1 = r'More than (?P<years_of_service1>[2][0]) years|Less than (?P<years_of_service2>[1]) year|(?P<years_of_service3>[0-9][0-9]?).0|(?P<years_of_service_lower>[0-9][0-9]?)-(?P<years_of_service_upper>[0-9][0-9]?)'
yos = combined_updated['institute_service'].str.extractall(pattern1)
yos = yos.astype('float64')
```

```python
yos.loc[(yos['years_of_service3'].isnull())&(yos['years_of_service1'].isnull())&(yos['years_of_service2'].isnull()),'years_of_service3'] = (yos.loc[yos['years_of_service3'].isnull(),'years_of_service_lower'] + yos.loc[yos['years_of_service3'].isnull(),'years_of_service_upper'])/2
yos.loc[yos['years_of_service2'].notnull(),'years_of_service3'] = yos.loc[yos['years_of_service2'].notnull(),'years_of_service2']
yos.loc[yos['years_of_service1'].notnull(),'years_of_service3'] = yos.loc[yos['years_of_service1'].notnull(),'years_of_service1']
combined_updated.loc[:,'institute_service'] = yos.loc[:,'years_of_service3'].droplevel('match')
```

```python
def career_stage(element):
    if pd.isnull(element):
        return np.nan
    elif element < 3:
        return 'New'
    elif (element >= 3) and (element < 6):
        return 'Experienced'
    elif (element >= 6) and (element < 10):
        return 'Established'
    else:
        return 'Veteran'
    
combined_updated['service_cat'] = combined_updated['institute_service'].apply(career_stage)
```

We have successfully created a `service_cat` column in `combined_updated` which categorizes the stage each employee is in their career at the corresponding institute based on their `institute_service` aka their years of service. This will help us delineate between newer and older employees when doing the analysis of the causes of their resignations.

```python
combined_updated['dissatisfied'].value_counts(dropna=False)
```

```python
summary = combined_updated.pivot_table(values='dissatisfied',index='service_cat')
```

```python
%matplotlib inline
summary.plot(kind='bar')
```

And so we have it! Now let's take a look back at our original questions
1. Are employees who only worked for the institutes for a short period of time resigning due to some kind of dissatisfaction? What about employees who have been there longer?

Data indicates that of the employees who worked for the institutes for only a short period (experienced and new employees) before they resigned, ~30% were dissatisfied when the resigned.

### 3.2 - Question 2

Next we need to answer the following question:

2. Are younger employees resigning due to some kind of dissatisfaction? What about older employees?

First let's take a look at the age spread of the employees.

```python
combined_updated['age'].value_counts(dropna=False)
```

Looks like there's quite a spread. Unfortunately, it also looks like the format of the age data is inconsistent. We'll need to play some clever regex tricks to extract the data we need.

```python
display(combined_updated['age'].shape)
pattern2 = r'(?P<age_lower1>[0-9][0-9])-(?P<age_upper1>[0-9][0-9])|(?P<age_lower2>[0-9][0-9])  (?P<age_upper2>[0-9][0-9])|(?P<age_lower3>[0-9][0-9]) or older|(?P<age_upper3>[0-9][0-9]) or younger'
ages = combined_updated['age'].str.extractall(pattern2)
ages['age'] = pd.Series(dtype='float64')
ages = ages.droplevel(1)
# ages['age'] = ages['age'].astype('float64')
for row in ages.index:
    if pd.notnull(ages.loc[row,'age_lower1']) and pd.notnull(ages.loc[row,'age_upper1']):
        ages.loc[row,'age'] = np.mean([float(ages.loc[row,'age_lower1']),float(ages.loc[row,'age_upper1'])])
    elif pd.notnull(ages.loc[row,'age_lower2']) and pd.notnull(ages.loc[row,'age_upper2']):
        ages.loc[row,'age'] = np.mean([float(ages.loc[row,'age_lower2']),float(ages.loc[row,'age_upper2'])])
    elif pd.notnull(ages.loc[row,'age_lower3']):
        ages.loc[row,'age'] = float(ages.loc[row,'age_lower3'])
    elif pd.notnull(ages.loc[row,'age_upper3']):
        ages.loc[row,'age'] = float(ages.loc[row,'age_upper3'])

combined_updated = combined_updated.drop(columns=['age'])
combined_updated = combined_updated.merge(ages['age'], how='left', left_index=True, right_index=True)
```

Now we have cleaned up the age column. Let's plot out the age distribution vs dissatisfaction.

```python
summary = combined_updated.pivot_table(values='dissatisfied',index='age')
summary.plot(kind='bar')
```

And so we have it (take 2)! Now we can finally answer the second question:

2. Are younger employees resigning due to some kind of dissatisfaction? What about older employees?

Data indicates that of the younger folks that resign, less than 50% are doing so due to dissatisfaction. However, the percentage of folks that are dissatisfied at the time they resign tends to increase with the age of the resigning employee
